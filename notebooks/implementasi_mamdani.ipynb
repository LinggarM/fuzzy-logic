```
import numpy as np
import skfuzzy as fzz
import matplotlib.pyplot as plt
%matplotlib inline
```

```
# range input
x_uan = np.arange(0,40,1)
x_raport = np.arange(0,100,1)
x_tpa = np.arange(0,100,1)
x_fisik = np.arange(0,100,1)
x_wawancara = np.arange(0,100,1)
x_psikologi = np.arange(0,100,1)

# range output
x_kelulusan = np.arange(0,100,1)
```

```
# Fungsi mambership
uan_low = fzz.trimf(x_uan,[0,10,20])
uan_mid = fzz.trimf(x_uan,[15,22,29])
uan_high = fzz.trimf(x_uan,[25,35,40])

raport_low = fzz.trimf(x_raport,[0,55,65])
raport_mid = fzz.trimf(x_raport,[60,72.5,85])
raport_high = fzz.trimf(x_raport,[75,88,100])

tpa_low = fzz.trimf(x_tpa,[0,50,60])
tpa_mid = fzz.trimf(x_tpa,[55,67.5,80])
tpa_high = fzz.trimf(x_tpa,[75,88,100])

fisik_low = fzz.trimf(x_fisik,[0,50,60])
fisik_mid = fzz.trimf(x_fisik,[55,70,85])
fisik_high = fzz.trimf(x_fisik,[75,85,100])

wawancara_low = fzz.trimf(x_wawancara,[0,65,79])
wawancara_high = fzz.trimf(x_wawancara,[70,80,100])

psikologi_low = fzz.trimf(x_psikologi,[0,65,75])
psikologi_high = fzz.trimf(x_psikologi,[70,75,100])

lulus = fzz.trimf(x_kelulusan,[60,75,100])
tidak_lulus = fzz.trimf(x_kelulusan,[0,52,70])
```

```
# grafik mambership function

fig, (ax0, ax1, ax2, ax3, ax4, ax5) = plt.subplots(nrows=6, figsize=(10,20))

ax0.plot(x_uan, uan_low, 'b', linewidth=1.5, label = 'rendah')
ax0.plot(x_uan, uan_mid, 'g', linewidth=1.5, label = 'sedang')
ax0.plot(x_uan, uan_high, 'r', linewidth=1.5, label = 'tinggi')
ax0.set_title('Nilai UAN')
ax0.legend()

ax1.plot(x_raport, raport_low, 'b', linewidth=1.5, label = 'rendah')
ax1.plot(x_raport, raport_mid, 'g', linewidth=1.5, label = 'sedang')
ax1.plot(x_raport, raport_high, 'r', linewidth=1.5, label = 'tinggi')
ax1.set_title('Nilai Raport')
ax1.legend()

ax2.plot(x_tpa, tpa_low, 'b', linewidth=1.5, label = 'rendah')
ax2.plot(x_tpa, tpa_mid, 'g', linewidth=1.5, label = 'sedang')
ax2.plot(x_tpa, tpa_high, 'r', linewidth=1.5, label = 'tinggi')
ax2.set_title('TPA')
ax2.legend()

ax3.plot(x_fisik, fisik_low, 'b', linewidth=1.5, label = 'kurang')
ax3.plot(x_fisik, fisik_mid, 'g', linewidth=1.5, label = 'cukup')
ax3.plot(x_fisik, fisik_high, 'r', linewidth=1.5, label = 'baik')
ax3.set_title('Tes Fisik')
ax3.legend()

ax4.plot(x_wawancara, wawancara_low, 'b', linewidth=1.5, label = 'kurang')
ax4.plot(x_wawancara, wawancara_high, 'r', linewidth=1.5, label = 'baik')
ax4.set_title('Tes Wawancara')
ax4.legend()

ax5.plot(x_psikologi, psikologi_low, 'b', linewidth=1.5, label = 'kurang')
ax5.plot(x_psikologi, psikologi_high, 'r', linewidth=1.5, label = 'baik')
ax5.set_title('Tes Psikologi')
ax5.legend()
```

```
# input
nilai_uan = 30.74
nilai_raport = 79.84
nilai_tpa = 83.33
nilai_fisik = 80
nilai_wawancara = 85
nilai_psikologi = 50

# Fuzzifikasi
input_uan = []
input_uan.append(fzz.interp_membership(x_uan,uan_low,nilai_uan))
input_uan.append(fzz.interp_membership(x_uan,uan_mid,nilai_uan))
input_uan.append(fzz.interp_membership(x_uan,uan_high,nilai_uan))

input_raport = []
input_raport.append(fzz.interp_membership(x_raport,raport_low,nilai_raport))
input_raport.append(fzz.interp_membership(x_raport,raport_mid,nilai_raport))
input_raport.append(fzz.interp_membership(x_raport,raport_high,nilai_raport))

input_tpa = []
input_tpa.append(fzz.interp_membership(x_tpa,tpa_low,nilai_tpa))
input_tpa.append(fzz.interp_membership(x_tpa,tpa_mid,nilai_tpa))
input_tpa.append(fzz.interp_membership(x_tpa,tpa_high,nilai_tpa))

input_fisik = []
input_fisik.append(fzz.interp_membership(x_fisik,fisik_low,nilai_fisik))
input_fisik.append(fzz.interp_membership(x_fisik,fisik_mid,nilai_fisik))
input_fisik.append(fzz.interp_membership(x_fisik,fisik_high,nilai_fisik))

input_wawancara = []
input_wawancara.append(fzz.interp_membership(x_wawancara,wawancara_low,nilai_wawancara))
input_wawancara.append(fzz.interp_membership(x_wawancara,wawancara_high,nilai_wawancara))

input_psikologi = []
input_psikologi.append(fzz.interp_membership(x_psikologi,psikologi_low,nilai_psikologi))
input_psikologi.append(fzz.interp_membership(x_psikologi,psikologi_high,nilai_psikologi))
```
